```python
# ===========================
# MarketPulse Stock Data Notebook (Colab-ready)
# ===========================

# Step 1: Install required packages
!pip install yfinance --quiet
!pip install pandas_datareader --quiet

# Step 2: Import libraries
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# ---------------------------
# Step 3: Fetch stock data
# ---------------------------
# Single ticker example
ticker = "AAPL"
df = yf.download(ticker, start="2015-01-01", end="2025-01-01")
df.reset_index(inplace=True)  # Make Date a column
print("Data sample:")
print(df.head())

# ---------------------------
# Step 4: Preprocessing
# ---------------------------
# Use only 'Close' price for LSTM
data = df[['Close']].values

# Normalize data
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(data)

# Create training dataset
training_data_len = int(len(scaled_data) * 0.8)
train_data = scaled_data[0:training_data_len]

# Split into sequences
def create_sequences(data, seq_length=60):
    x = []
    y = []
    for i in range(seq_length, len(data)):
        x.append(data[i-seq_length:i, 0])
        y.append(data[i, 0])
    return np.array(x), np.array(y)

seq_length = 60
X_train, y_train = create_sequences(train_data, seq_length)

# Reshape for LSTM: [samples, time steps, features]
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))

# ---------------------------
# Step 5: Build LSTM Model
# ---------------------------
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1],1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')

# ---------------------------
# Step 6: Train the model
# ---------------------------
model.fit(X_train, y_train, batch_size=32, epochs=5)

# ---------------------------
# Step 7: Create test dataset
# ---------------------------
test_data = scaled_data[training_data_len - seq_length:]
X_test, y_test = create_sequences(test_data, seq_length)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1],1))

# Predict
predictions = model.predict(X_test)
predictions = scaler.inverse_transform(predictions)  # Rescale to original

# ---------------------------
# Step 8: Plot results
# ---------------------------
train = df[:training_data_len]
valid = df[training_data_len:]
valid['Predictions'] = predictions

plt.figure(figsize=(14,6))
plt.plot(train['Close'], label='Train Data')
plt.plot(valid['Close'], label='Actual Price')
plt.plot(valid['Predictions'], label='Predicted Price')
plt.title(f"{ticker} Stock Price Prediction")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.show()
```

```python
%%writefile app.py
import streamlit as st
import yfinance as yf
import numpy as np
import pandas as pd
from tensorflow.keras.models import load_model
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Load trained model (make sure it's saved in Colab)
model = load_model("marketpulse_model.h5")

st.title("MarketPulse: Stock Price Prediction")

# User inputs
ticker = st.text_input("Enter stock ticker:", "AAPL")
start_date = st.date_input("Start date", pd.to_datetime("2020-01-01"))
end_date = st.date_input("End date", pd.to_datetime("2025-01-01"))

if st.button("Predict"):
    # Fetch stock data
    df = yf.download(ticker, start=start_date, end=end_date)
    df1 = df['Close']

    # Scale
    scaler = MinMaxScaler(feature_range=(0,1))
    df1_scaled = scaler.fit_transform(np.array(df1).reshape(-1,1))

    # Prepare last 60 days for prediction
    last_60_days = df1_scaled[-60:]
    X_test = np.reshape(last_60_days, (1, last_60_days.shape[0], 1))

    # Predict next day
    pred_scaled = model.predict(X_test)
    pred = scaler.inverse_transform(pred_scaled)

    st.write(f"Predicted next day Close price for {ticker}: ${pred[0][0]:.2f}")

    # Plot historical data
    st.line_chart(df1)
```
